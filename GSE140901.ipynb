```python
import os
import pandas as pd
def read_matrix(file):
    with open(file)as f:
        result = {}
        for line in f:
            lines = line.strip('\n').replace('"','').split('\t')
            if len(lines[0])>0:
                if lines[0][1:] in result:
                    result['_'.join([lines[0][1:],lines[1].split(':')[0]])] = lines[1:]
                else:
                    result[lines[0][1:]] = lines[1:] 
    sample_len =max([len(i) for i in result.values()])             
    return result,sample_len

def dic2df(dic:dict,sample_len):
    df = pd.DataFrame()
    for k in dic:
        if len(dic[k])==sample_len:
            df[k] = dic[k]
    return df 
```

```python
test_file = 'new_data/GSE140901/GSE140901_series_matrix.txt'
pd.set_option('display.max_columns', None)
result,sample_len = read_matrix(test_file)
df_matrix = dic2df(result,sample_len)
df_matrix[:5]
```
