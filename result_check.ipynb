```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 汇总所有细胞类型的前10个重要特征
topn = 20
top_features_df = []

for celltype, shap_df in feature_importance_dic.items():
    top_features = shap_df.head(topn).copy()
    top_features["Cell_Type"] = celltype
    top_features_df.append(top_features)

combined_df = pd.concat(top_features_df, axis=0)

# 可视化：热图方式呈现重要性
pivot_df = combined_df.pivot_table(index='Cell_Type', columns='Feature', values='Mean SHAP Value', fill_value=0)

plt.figure(figsize=(15, 6))
sns.heatmap(pivot_df, cmap='viridis',  )
plt.title("Top 20 Features SHAP Importance Across Cell Types")

plt.tight_layout()
plt.savefig('/mnt/data/wangw/test/Top 20 Features SHAP Importance Across Cell Types.pdf')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
for cell_type, shap_genes in top_gene_sets.items():
    print(f"处理细胞类型: {cell_type}，SHAP基因数: {len(shap_genes)}")

    # 只选该类型的细胞
    mask = adata.obs['major_cluster'] == cell_type
    adata_sub = adata[mask].copy()

    # 只保留在表达矩阵中存在的 SHAP 基因
    valid_genes = [g for g in shap_genes if g in adata.var_names]
    if len(valid_genes) == 0:
        print(f"{cell_type} 中无合法 SHAP 基因，跳过")
        continue

    # 构建表达 DataFrame（N cells × SHAP genes）
    expr_df = pd.DataFrame(adata_sub[:, valid_genes].X.toarray(),
                           columns=valid_genes,
                           index=adata_sub.obs_names)

    # 合并 TCR 信息
    merged = pd.concat([adata_sub.obs, expr_df], axis=1)

    # 遍历 top 5 SHAP 基因绘制 violin 图（克隆 vs 非克隆）
    top5 = valid_genes[:5]
    for gene in top5:
        plt.figure(figsize=(5, 4))
        sns.violinplot(data=merged, x='is_clonal', y=gene)
        plt.title(f"{gene} expression in clonal vs non-clonal {cell_type} cells")
        plt.tight_layout()
        plt.show()
```

```python
adata.obs['Cell_Type'] = adata.obs['sub_cluster'].apply(lambda x:x.split('_')[0])
adata.obs['Cell_Type']
for cell_type, shap_genes in top_gene_sets.items():
    if cell_type != 'CD8':
        continue
    print(f"开始处理细胞类型: {cell_type}")

    adata_sub = adata[adata.obs['Cell_Type'] == cell_type].copy()
    print(adata_sub.shape)
    valid_genes = [g for g in shap_genes if g in adata.var_names]
    print(f"  SHAP基因数: {len(valid_genes)}")

    if len(valid_genes) == 0:
        # print(f"  跳过：没有合法基因")
        continue

    expr_df = pd.DataFrame(adata_sub[:, valid_genes].X.toarray(),
                           columns=valid_genes,
                           index=adata_sub.obs_names)

    merged = pd.concat([adata_sub.obs, expr_df], axis=1)

    # 检查 clone.id
    print(f"  有 clone.id 的细胞数：{merged['clone.id'].notna().sum()}")

    df_clone = merged.groupby('clone.id')[valid_genes].mean().dropna()
    print(f"  clone × gene 表 shape: {df_clone.shape}")

    if df_clone.shape[0] == 0 or df_clone.shape[1] == 0:
        # print(f"  跳过：没有足够数据")
        continue

    df_plot = df_clone.head(30)
    print(f"  画图数据 shape: {df_plot.shape}")

    cg = sns.clustermap(df_plot, cmap='viridis', figsize=(10, 6))
    cg.fig.suptitle(f"{cell_type}: SHAP gene expression by clone.id")
    cg.savefig(f"heatmap_{cell_type}.png")
    plt.show()
```

```python
import os
import pickle
file ='model_reult.pickle'

model_reult = pickle.load(open(file,'rb'))
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,auc,
    roc_curve,confusion_matrix)
from sklearn.model_selection import StratifiedKFold, train_test_split,RandomizedSearchCV
# 设置图形

plt.rcParams['font.family'] = 'Arial'
# 遍历所有模型
def compute_roc_auc_ci(y_true, y_scores, n_bootstraps=1000, alpha=0.95):
    rng = np.random.RandomState(42)
    bootstrapped_scores = []
    # y_true = y_true.to_numpy() if isinstance(y_true, pd.Series) else y_true
    y_true = np.array(y_true)
    for _ in range(n_bootstraps):
        indices = rng.randint(0, len(y_scores), len(y_scores))
        if len(np.unique(y_true[indices])) < 2:
            continue
        score = roc_auc_score(y_true[indices], y_scores[indices])
        bootstrapped_scores.append(score)

    sorted_scores = np.array(bootstrapped_scores)
    sorted_scores.sort()
    lower = sorted_scores[int((1.0 - alpha) / 2 * len(sorted_scores))]
    upper = sorted_scores[int((1.0 + alpha) / 2 * len(sorted_scores))]
    return lower, upper
for cell in model_reult:
    plt.figure(figsize=(10,8))
    for model_name in model_reult['all']:
        acc, auc_val, fpr, tpr, model, X_test, adata_test, le, metrics_list, types_clean, dflist = model_reult['all'][model_name]
        
        # 如果没有fpr和tpr，跳过
        if fpr is None or tpr is None:
            continue
        
        # 重新计算AUC以防 auc_val 缺失或不准
        auc_score = auc(fpr, tpr)

        
        plt.plot(fpr, tpr, lw=2, label=f"{model_name}: AUC: {auc_score:.3f} ")

    # 画对角线（随机分类器参考线）
    plt.plot([0,1], [0,1], color="gray", lw=1.5, linestyle="--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])

    # 设置图形属性
    plt.xlabel("False Positive Rate (1 - Specificity)")
    plt.ylabel("True Positive Rate (Sensitivity)")
    plt.title(f"ROC Curves of Different Models : {cell} ")
    plt.legend(loc="lower right", frameon=False)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    
    plt.savefig(f'Data_Visualization/ROC_Curves_of_Different_Models:{cell}.pdf')
    plt.show()
    plt.close()
```
