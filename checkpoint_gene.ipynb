```python
import pandas as pd
import numpy as np

# 所有基因集合
checkpoint_genes = [
    # 抑制型
    "PDCD1", "CD274", "PDCD1LG2", "CTLA4", "LAG3", "HAVCR2", "TIGIT", "VTCN1", "IDO1",
    # 激活型
    "CD28", "ICOS", "CD40", "TNFRSF9", "TNFRSF18", "TNFRSF4", "CD27",'CD48',
    # 可选拓展
    "FOXP3", "HLA-A", "HLA-B", "HLA-C", "HLA-DRA", "HLA-DRB1", "B2M",
    "IFNG", "IL10", "IL2", "IL6", "CXCL9", "CXCL10", "CXCL13",
    "STAT1", "IRF1",'BTN3A1','BTN2A2','LAIR1'
]

# 添加 group 列
adata.obs["group"] = adata.obs["pre_post"].astype(str) + "_" + adata.obs["response"].astype(str)

# 筛选存在于 var_names 中的基因
genes_in_data = [g for g in checkpoint_genes if g in adata.var_names]

# 创建一个存储结果的列表
summary_list = []

# 对每个 gene 和每个 group 进行遍历
for gene in genes_in_data:
    # 获取 log1p 表达值
    expr = adata[:, gene].X.toarray().flatten() if hasattr(adata.X, "toarray") else adata[:, gene].X.flatten()
    expr_log = np.log1p(expr)
    
    # 构建总表达统计
    total_stats = {
        "gene": gene,
        "group": "ALL",
        "min": np.min(expr_log),
        "max": np.max(expr_log),
        "mean": np.mean(expr_log),
        "std": np.std(expr_log)
    }
    summary_list.append(total_stats)

    # 每个 group 的统计
    for g in adata.obs["group"].unique():
        idx = (adata.obs["group"] == g).values
        expr_g = expr_log[idx]
        group_stats = {
            "gene": gene,
            "group": g,
            "min": np.min(expr_g),
            "max": np.max(expr_g),
            "mean": np.mean(expr_g),
            "std": np.std(expr_g)
        }
        summary_list.append(group_stats)

# 转为 DataFrame
summary_df = pd.DataFrame(summary_list)

summary_df
```

```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

def analyze_significance_by_gene_group(adata, checkpoint_genes, filename="immune_checkpoint_significance.csv"):
    # 定义 group 两两比较组合
    group_pairs = [
        ("pre_responder", "pre_non-responder"),
        ("post_responder", "post_non-responder"),
        ("pre_responder", "post_responder"),
        ("pre_non-responder", "post_non-responder"),
    ]

    # 标注显著性
    def get_sig(p):
        if p < 0.001: return '***'
        elif p < 0.01: return '**'
        elif p < 0.05: return '*'
        else: return 'ns'

    # 准备结果列表
    records = []

    # 添加 group 列
    adata.obs["group"] = adata.obs["pre_post"].astype(str) + "_" + adata.obs["response"].astype(str)

    # 遍历每个细胞类型
    for cell in sorted(set(adata.obs["cell_type"])):
        adata_cell = adata[adata.obs["cell_type"] == cell].copy()

        for gene in checkpoint_genes:
            if gene not in adata_cell.var_names:
                continue

            # 获取表达并 log1p 转换
            expr = adata_cell[:, gene].X.toarray().flatten() if hasattr(adata_cell.X, "toarray") else adata_cell[:, gene].X.flatten()
            expr_log = np.log1p(expr)
            group_series = adata_cell.obs["group"]

            for g1, g2 in group_pairs:
                idx1 = group_series == g1
                idx2 = group_series == g2
                if idx1.sum() == 0 or idx2.sum() == 0:
                    continue

                vals1 = expr_log[idx1]
                vals2 = expr_log[idx2]
                vals3 = expr[idx1]
                vals4 = expr[idx2]
                vals3_nonzero = vals3[vals3 != 0]
                vals4_nonzero = vals4[vals4 != 0]
                if len(vals3_nonzero) == 0:
                    vals3_nonzero =  vals3
                if len(vals4_nonzero) == 0:
                    vals4_nonzero =  vals4
                # 计算统计量
                stat, p = mannwhitneyu(vals1, vals2, alternative="two-sided")
                mean1, mean2 = np.mean(vals1), np.mean(vals2)
                mean3, mean4 = np.mean(vals3), np.mean(vals4)
                mean3_nozero, mean4_nozero = np.mean(vals3_nonzero), np.mean(vals4_nonzero)
                log2fc = np.log2(mean1 + 1e-6) - np.log2(mean2 + 1e-6)
                stat, p_ = mannwhitneyu(vals3, vals4, alternative="two-sided")
                records.append({
                    "cell_type": cell,
                    "gene": gene,
                    "group1": g1,
                    "group2": g2,
                    "mean1_log1p": mean1,
                    "mean2_log1p": mean2,
                    "mean1_exp": mean3,
                    "mean2_exp": mean4,
                    "mean1_exp_nozero": mean3_nozero,
                    "mean2_exp_nozero": mean4_nozero,
                    "log2FC": log2fc,
                    "p_value": p,
                    "p_value_nozero": p_,
                    "significance": get_sig(p),
                    "significance_nozero": get_sig(p_)
                })
    adata_cell = adata.copy()
    cell = 'all'
    for gene in checkpoint_genes:
        if gene not in adata_cell.var_names:
            continue

        # 获取表达并 log1p 转换
        expr = adata_cell[:, gene].X.toarray().flatten() if hasattr(adata_cell.X, "toarray") else adata_cell[:, gene].X.flatten()
        expr_log = np.log1p(expr)
        group_series = adata_cell.obs["group"]

        for g1, g2 in group_pairs:
            idx1 = group_series == g1
            idx2 = group_series == g2
            if idx1.sum() == 0 or idx2.sum() == 0:
                continue

            vals1 = expr_log[idx1]
            vals2 = expr_log[idx2]
            vals3 = expr[idx1]
            vals4 = expr[idx2]
            vals3_nonzero = vals3[vals3 != 0]
            vals4_nonzero = vals4[vals4 != 0]
            if len(vals3_nonzero) == 0:
                vals3_nonzero =  vals3
            if len(vals4_nonzero) == 0:
                vals4_nonzero =  vals4
            # 计算统计量
            stat, p = mannwhitneyu(vals1, vals2, alternative="two-sided")
            mean1, mean2 = np.mean(vals1), np.mean(vals2)
            mean3, mean4 = np.mean(vals3), np.mean(vals4)
            mean3_nozero, mean4_nozero = np.mean(vals3_nonzero), np.mean(vals4_nonzero)
            log2fc = np.log2(mean1 + 1e-6) - np.log2(mean2 + 1e-6)
            stat, p_ = mannwhitneyu(vals3, vals4, alternative="two-sided")
            records.append({
                "cell_type": cell,
                "gene": gene,
                "group1": g1,
                "group2": g2,
                "mean1_log1p": mean1,
                "mean2_log1p": mean2,
                "mean1_exp": mean3,
                "mean2_exp": mean4,
                "mean1_exp_nozero": mean3_nozero,
                "mean2_exp_nozero": mean4_nozero,
                "log2FC": log2fc,
                "p_value": p,
                "p_value_nozero": p_,
                "significance": get_sig(p),
                "significance_nozero": get_sig(p_)
            })
    # 保存为 CSV
    df_result = pd.DataFrame(records)
    _, pvals_adj, _, _ = multipletests(df_result["p_value"].values, method="fdr_bh")
    df_result["adj_p_value"] = pvals_adj
    df_result["adj_significance"] = df_result["adj_p_value"].apply(get_sig)
    df_result.to_csv(filename, index=False)
    print(f"✅ 显著性分析结果已保存到: {filename}")
    return df_result
df_result = analyze_significance_by_gene_group(adata, checkpoint_genes, filename="immune_checkpoint_significance.csv")
```
